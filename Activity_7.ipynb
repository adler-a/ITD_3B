```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')

import folium
from folium.plugins import HeatMap
```

```python
chi_crimes = pd.read_csv('datasets//Chicago_Crimes.csv')
chi_crimes['Date'] = pd.to_datetime(chi_crimes['Date'], errors='coerce')
```

```python
chi_crimes
```

```python
chi_crimes.columns
```

<h1>THEFT ANALYSIS</h1>

```python
theft_filter = chi_crimes[chi_crimes['Primary Type'] == 'THEFT'] #59k
```

```python
theft_filter2 = theft_filter[theft_filter['Location Description'] == 'SIDEWALK'] #1.6k
```

```python
theft_final = theft_filter2.groupby(['Latitude', 'Longitude']).size().reset_index(name= 'theft_count') #1.4k
```

```python
theft_heatmap_data = theft_final[['Latitude','Longitude','theft_count']].values.tolist()
```

```python
theft_my_map =  folium.Map(location = [40,-85], zoom_start = 6)
HeatMap(theft_heatmap_data).add_to(theft_my_map)
```

```python
theft_my_map.save('theft.html')
print('data save to theft.html')
theft_my_map
```
